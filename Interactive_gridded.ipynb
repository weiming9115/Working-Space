```python
import sys
import os 
import matplotlib.pyplot as plt
import holoviews as hv
import geoviews as gv
import geoviews.feature as gf
import numpy as np
import xarray as xr
from datetime import datetime
from cartopy import crs

gv.extension('bokeh', 'matplotlib')
```

```python
# import gridded data: NCEP reanalysis uwind
dataset=xr.open_dataset('uwnd.day.2010_2016.45NS.nc')
dataset
```

```python
ds=dataset.sel(time=datetime(2016,7,1),lon=slice(0,360),lat=slice(-45,45)).u
ds1=dataset.sel(time=datetime(2016,7,1),lon=slice(0,360),lat=slice(-45,45),lev=200).u
uwnd=gv.Dataset(ds);
uwnd # declaring the variable for holoviews
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeat

fig = plt.figure(figsize=(14,8))

ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180)) # set up central of map
us=xr.plot.contourf(ds1,transform=ccrs.PlateCarree(),cmap='jet',robust='true',add_colorbar=False)
ax.coastlines()
ax.add_feature(cfeat.LAND,zorder=0,edgecolor='k') # maskout LAND
cbar=plt.colorbar(us,fraction=0.012, pad=0.025);
cbar.set_label('[m/s]',fontsize=12)
plt.title('NCEP Reanalysis Uwind, 200hap, 45S-45N, July 1,2016',fontsize=14)
plt.xlabel('Logitude',fontsize=12);plt.ylabel('Latitude',fontsize=12)
plt.xticks(np.linspace(-180,180,13),['0','30E','60E','90E','120E','150E','180','150W','120W','90w'
                                    ,'60W','30W'])
plt.yticks(np.linspace(-45,45,7),['45S','30S','15S','EQ','15N','30N','45N'])
                                    
plt.show()
```
