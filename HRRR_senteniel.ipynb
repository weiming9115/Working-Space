```python
import sys
from datetime import datetime
import os
import matplotlib.pyplot as plt
import xarray as xr
import geoviews as gv
import numpy as np
import cartopy.crs as ccrs
import cartopy.feature as cfeat
import pygrib 

gv.extension('bokeh', 'matplotlib')
```

```python
os.chdir('/Users/willy/Desktop/UMworks/HRRR_PW_6hourly/')
grbs = pygrib.open('hrrr.t18z.20180412.grib2')  
grb = grbs.read(1)[0] # read returns a list with the next N (N=1 in this case) messages
grb
```

```python
# reading grib2 data
lats, lons = grb.latlons() 
pw = grb.values
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection=ccrs.LambertConformal(central_latitude = 25, 
                             central_longitude = 265, 
                             standard_parallels = (25, 25)))

plt.pcolormesh(lons,lats,pw, transform = ccrs.PlateCarree(),cmap='jet')
ax.coastlines('50m')
ax.add_feature(cfeat.STATES)
h=plt.colorbar(fraction=0.02, pad=0.025)
h.set_label('Precipitable water [mm]',fontsize=14)
#fig.savefig('HRRR_PW_0412.png',facecolor='w',bbox_inches='tight')

plt.show()
```

```python
#date_sel=['hrrr.t18z.20180412.grib2','hrrr.t18z.20180424.grib2','hrrr.t18z.20180506.grib2',
#          'hrrr.t18z.20180518.grib2','hrrr.t18z.20180530.grib2','hrrr.t18z.20180611.grib2',
#          'hrrr.t18z.20180623.grib2']
# Precipitable water difference of two dates
# reading grib2 data
grbs = pygrib.open('hrrr.t12z.20180412.grib2')  
grb = grbs.read(1)[0] # read returns a list with the next N (N=1 in this case) messages
pw = grb.values
grbs2 = pygrib.open('hrrr.t12z.20180424.grib2')  
grb2 = grbs2.read(1)[0] # read returns a list with the next N (N=1 in this case) messages
lats, lons = grb2.latlons() 
pw2 = grb2.values

fig = plt.figure(figsize=(8,10))
ax = plt.axes(projection=ccrs.LambertConformal(central_latitude = 25, 
                             central_longitude = 265, 
                             standard_parallels = (25, 25)))

plt.pcolormesh(lons,lats,pw2-pw, transform = ccrs.PlateCarree(),cmap='jet')
ax.coastlines()
ax.add_feature(cfeat.STATES)
h=plt.colorbar(fraction=0.02, pad=0.025)
h.set_label('Precipitable water [mm]',fontsize=14)
plt.title('$\Delta$'+'PW'+'  '+'[2018-04-24] - [2018-04-12]',fontsize=18)


fig = plt.figure(figsize=(6,15))
ax = plt.axes(projection=ccrs.LambertConformal(central_latitude = 25, 
                             central_longitude = 265, 
                             standard_parallels = (25, 25)))
ax.set_extent([-123.75, -122, 44.75,46])
plt.pcolormesh(lons,lats,pw2-pw, transform = ccrs.PlateCarree(),cmap='jet',vmin=-10,vmax=10)
ax.coastlines()
ax.add_feature(cfeat.STATES)
h=plt.colorbar(fraction=0.05, pad=0.03)
h.set_label('$\Delta$'+' Precipitable Water [mm]',fontsize=16)
plt.title('NOAA_HRRR, 20180424 - 20180412 \n lat=[44.75,46],lon=[-123.75,-122]',fontsize=14)
ax.set_aspect(1.6)

plt.show()

#fig.savefig('HRRR_PWDIFF_0424_0412_JET.png', dpi=None, facecolor='w',bbox_inches='tight')
```
